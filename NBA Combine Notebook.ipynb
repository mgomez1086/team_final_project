```python
url = 'https://stats.nba.com/draft/combine-spot-up/'
browser.visit(url)
```

```python
# visiting browser page
html = browser.html

# setting up beautiful soup
soup = BeautifulSoup(html, 'html.parser')

my_table = soup.find('table')
my_table
```

```python
list_colnames=[]

for each_col in my_table.find('thead').findAll('th'):
    list_colnames.append(each_col.text)
    # print(each_col.text)
```

```python
list_colnames
```

```python
list_row = []
for x in range(1, 107):
    for each_row in my_table.find('tbody').findAll('tr'):
        list_each_col_per_row = []
        for each_col_per_row in each_row.findAll('td'):
        
            list_each_col_per_row.append(each_col_per_row.text)
        #print(each_col_per_row.text)
        list_row.append(list_each_col_per_row)
        browser.find_by_css('option').first.click()
    #print('----')
```

```python
list_row
```

```python
df = pd.DataFrame(list_row, columns = list_colnames)
```
